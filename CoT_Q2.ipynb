# 4. Convert Date and Time- cleaning

```python
# Convert 'date_of_infraction' to datetime format, coercing errors to NaT
data['date_of_infraction'] = pd.to_datetime(data['date_of_infraction'], format='%Y%m%d', errors='coerce')

# Check for rows where the conversion failed
invalid_dates = data[data['date_of_infraction'].isna()]

# Display invalid dates
print(f"Number of invalid dates: {len(invalid_dates)}")
invalid_dates.head()
```

```python
# Convert 'date_of_infraction' to datetime format
data['date_of_infraction'] = pd.to_datetime(data['date_of_infraction'], format='%Y%m%d')

# Extract useful date parts
data['year'] = data['date_of_infraction'].dt.year
data['month'] = data['date_of_infraction'].dt.month
data['day_of_week'] = data['date_of_infraction'].dt.day_name()
```

```python
# Check for non-numeric values in 'time_of_infraction'
invalid_times = data[~data['time_of_infraction'].apply(lambda x: str(x).isdigit())]

# Display the invalid values
print(f"Number of invalid time values: {len(invalid_times)}")
print(invalid_times['time_of_infraction'].unique())
```

```python
# Check for any non-numeric values in the 'time_of_infraction' column
non_numeric_times = data[~data['time_of_infraction'].apply(lambda x: str(x).replace('.', '').isdigit())]

# Display the non-numeric values
print(non_numeric_times['time_of_infraction'].unique())
```

```python
# Drop rows where 'time_of_infraction' contains non-numeric values (like 'time_of_infraction')
data = data[data['time_of_infraction'].apply(lambda x: str(x).replace('.', '').isdigit())]
```

```python
# Remove the decimal part by converting to integer (since the numbers represent HHMM)
data['time_of_infraction'] = data['time_of_infraction'].astype(float).astype(int)

# Filter out invalid times (i.e., times where the hour part is greater than 23 or minute part is greater than 59)
valid_times = data['time_of_infraction'].apply(lambda x: 0 <= x // 100 <= 23 and 0 <= x % 100 <= 59)
```

```python
# Keep only valid times
data = data[valid_times]

# Convert the valid HHMM integers into time format
data['time_of_infraction'] = pd.to_datetime(data['time_of_infraction'].astype(str).str.zfill(4), format='%H%M').dt.time

# Check the first few rows to confirm the conversion
data[['time_of_infraction']].head()
```

# 6. Analyze Time-Based Patterns

```python
# Tickets by day of the week
day_of_week_counts = data['day_of_week'].value_counts()

# Tickets by hour of day (we need to extract the hour first)
data['hour_of_infraction'] = pd.to_datetime(data['time_of_infraction'], format='%H:%M:%S', errors='coerce').dt.hour
hour_of_day_counts = data['hour_of_infraction'].value_counts().sort_index()

# Display the counts
print(day_of_week_counts)
print(hour_of_day_counts)
```

# 7. Visualize the Results


```python
# Plot tickets by day of the week
plt.figure(figsize=(8,5))
sns.barplot(x=day_of_week_counts.index, y=day_of_week_counts.values, palette="viridis")
plt.title('Parking Tickets by Day of the Week')
plt.ylabel('Number of Tickets')
plt.xlabel('Day of Week')
plt.show()

# Plot tickets by hour of the day
plt.figure(figsize=(8,5))
sns.barplot(x=hour_of_day_counts.index, y=hour_of_day_counts.values, palette="magma")
plt.title('Parking Tickets by Hour of the Day')
plt.ylabel('Number of Tickets')
plt.xlabel('Hour of Day')
plt.show()
```
